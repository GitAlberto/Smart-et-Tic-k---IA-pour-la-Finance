## Extraction du texte

```python
# Importation des librairies
import os
import easyocr
import pandas as pd
import re
```

```python
# --- CONFIGURATION ---
dossier_images = "dataset_augmente"
fichier_sortie = "textes_bruts_tickets.csv"

print("Chargement de l'OCR avec les détails de position (detail=1)...")
reader = easyocr.Reader(['fr', 'en'], gpu=True) 

donnees_globales = []

if not os.path.exists(dossier_images):
    print(f"Dossier {dossier_images} introuvable.")
    exit()

print("Début de l'extraction spatiale...")

for fichier in os.listdir(dossier_images):
    if fichier.lower().endswith(('.png', '.jpg', '.jpeg')):
        chemin_image = os.path.join(dossier_images, fichier)
        print(f"  -> Traitement : {fichier}")
        
        # detail=1 renvoie : (Boîte_englobante, Texte, Confiance)
        resultats = reader.readtext(chemin_image, detail=1)
        
        for bbox, texte, confiance in resultats:
            texte = texte.strip()
            if len(texte) < 2:
                continue # On ignore direct les caractères tout seuls
                
            # Calcul du centre Y du texte (pour aligner les prix avec les produits plus tard)
            # bbox = [[x_haut_gauche, y_haut_gauche], [x_haut_droit, y_haut_droit], ...]
            y_haut = bbox[0][1]
            y_bas = bbox[3][1]
            y_centre = (y_haut + y_bas) / 2
            
            # Calcul du X gauche (pour savoir si c'est écrit à gauche ou à droite du ticket)
            x_gauche = bbox[0][0]

            # Détection basique : est-ce que ça ressemble à un prix ? (ex: 2.50, 14,99)
            est_un_prix = bool(re.search(r'\d+[.,]\d{2}', texte))
            
            donnees_globales.append({
                "nom_fichier": fichier,
                "texte_brut": texte,
                "y_centre": round(y_centre, 2), # La hauteur sur le ticket
                "x_gauche": round(x_gauche, 2), # La position horizontale
                "est_un_prix": est_un_prix,
                "confiance_ocr": round(confiance, 2),
                "categorie_cible": "" # À remplir plus tard pour l'IA
            })

# Sauvegarde du fichier
df = pd.DataFrame(donnees_globales)
df.to_csv(fichier_sortie, index=False, encoding='utf-8')

print("-" * 30)
print(f"Extraction des textes terminée !")
print(f"Fichier généré : {fichier_sortie}")
```
