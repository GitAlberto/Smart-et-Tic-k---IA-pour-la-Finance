## Data Augmentation

```python
import os
import cv2
import albumentations as A
import random

# --- CONFIGURATION ---
input_folder = "Dataset brute"
output_folder = "dataset_augmente"

# On vÃĐrifie que les dossiers existent
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# --- DÃFINITION DES TRANSFORMATIONS ---
# C'est la "recette" pour modifier les images
transform = A.Compose([
    # 1. Rotation alÃĐatoire entre -10 et +10 degrÃĐs (simule un scan de travers)
    A.Rotate(limit=10, p=0.7, border_mode=cv2.BORDER_REPLICATE),

    # 2. Changement alÃĐatoire de luminositÃĐ et contraste (simule un mauvais ÃĐclairage)
    A.RandomBrightnessContrast(brightness_limit=0.2, contrast_limit=0.2, p=0.7),

    # 3. Ajout de bruit ou de flou lÃĐger (simule une mauvaise camÃĐra)
    A.OneOf([
        A.GaussNoise(var_limit=(10.0, 50.0)),
        A.GaussianBlur(blur_limit=3),
        A.MotionBlur(blur_limit=3),
    ], p=0.5),
])

# --- EXÃCUTION ---
print(f"DÃĐbut de l'augmentation depuis '{input_folder}' vers '{output_folder}'...")
count = 0

# On parcourt toutes les images du dossier brut
for filename in os.listdir(input_folder):
    if filename.lower().endswith(('.png', '.jpg', '.jpeg')):
        image_path = os.path.join(input_folder, filename)

        # 1. Lecture de l'image avec OpenCV
        image = cv2.imread(image_path)
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB) # Convertir en RGB

        # 2. On sauvegarde d'abord l'originale dans le nouveau dossier
        original_name = f"orig_{filename}"
        cv2.imwrite(os.path.join(output_folder, original_name), cv2.cvtColor(image, cv2.COLOR_RGB2BGR))
        count += 1

        # 3. On gÃĐnÃĻre 3 variations par image
        for i in range(3):
            # Appliquer la transformation
            augmented = transform(image=image)
            augmented_image = augmented["image"]

            # Reconvertir en BGR pour OpenCV avant sauvegarde
            augmented_image_bgr = cv2.cvtColor(augmented_image, cv2.COLOR_RGB2BGR)

            # GÃĐnÃĐrer un nouveau nom (ex: ticket1_aug_0.jpg)
            base_name = os.path.splitext(filename)[0]
            extension = os.path.splitext(filename)[1]
            new_filename = f"{base_name}_aug_{i}{extension}"
            output_path = os.path.join(output_folder, new_filename)

            # Sauvegarder
            cv2.imwrite(output_path, augmented_image_bgr)
            count += 1
            print(f"  -> GÃĐnÃĐrÃĐ : {new_filename}")

print("-" * 30)
print(f"TerminÃĐ ! Nous avons maintenant {count} images dans '{output_folder}'.")
```
